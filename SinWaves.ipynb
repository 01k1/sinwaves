```python
import ipywidgets as widgets
from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np

%matplotlib nbagg

x = np.linspace(0, 2, 1000)
fig, ax = plt.subplots(1, figsize = (10, 4))
plt.suptitle('Sine Wave')


def update_plot(amp, freq, phase,play):
    ''' This function is linked to the slider and it replots the sine wave when the slider are changed.'''
    ax.clear()
    units = 'amp = {} $(psi)$\n phase = {} $(s)$ \n freq ={} $(Hz)$'
    y = amp * np.sin(2*np.pi*(freq*x-phase))
    ax.plot(x, y, label=units.format(amp, freq, play, phase))
    ax.legend(loc=1)
    ax.set_xlabel('$(s)$')
    ax.set_ylabel('$(psi)$')
    ax.set_xlim(x[0], x[-1])
    plt.show()
    
    
amp = widgets.FloatSlider(min = 1, max = 10, value = 1, description = 'Amp:')
phase = widgets.FloatSlider(min = 1, max = 10, value = 1, description = 'Phase:')



play = widgets.Play(min = 0, max = 20, value = 1, step = 1)
freq = widgets.FloatSlider(description = 'Freq:')
widgets.jslink((play, 'value'), (freq, 'value'))
widgets.HBox([play, freq])


widgets.interactive(update_plot, amp= amp, freq= freq, play = play, phase= phase)
```
